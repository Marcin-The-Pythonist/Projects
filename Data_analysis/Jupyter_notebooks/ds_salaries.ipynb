```python
# Directory & Libraries
%cd C:\Users\48505\Documents\Python\Notebooks\csv
import pandas as pd 
import numpy as np
```

```python
# Loading data
df = pd.read_csv("ds_salaries.csv")
df.head(5)
```

```python
# Cleaning data
df.isna().sum() #0
df.isnull().sum() #0
df.duplicated().sum() #0
df["employment_type"] = df["employment_type"].str.replace("PT","Part-time")
df["employment_type"] = df["employment_type"].str.replace("CT","Contract")
df["employment_type"] = df["employment_type"].str.replace("FL","Freelance")

df.remote_ratio = df.remote_ratio.astype("int")

df["remote_ratio"].mask((df["remote_ratio"] > 0) & (df["remote_ratio"] < 100) ,"Hybrid",inplace=True)
df["remote_ratio"] = df["remote_ratio"].replace(100,"Remote")
df["remote_ratio"] = df["remote_ratio"].replace(0,"On-site") 

df.remote_ratio.value_counts()

```

```python
# Exporting the clean data
df.to_csv("ds_salaries_clean.csv")
df.head(5)
```
