```python
# Cleaning data 
df["Rating"].fillna(value=0, inplace=True) #Replace NaN rating with 0.
df.dropna(axis=0) # Drop rows with NaN.
df.isna().sum()
print(df.shape)
df_clean = df.dropna(axis=0)
df_clean.isna().sum()
df_clean["Installs"] = df_clean["Installs"].str.replace("+","") # Get rid of '+' sign. 
df_clean
```
